```python
# ================================
# GPU CONFIGURATION (Kaggle)
# ================================
import tensorflow as tf
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        # Enable memory growth on all GPUs for safer usage.
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        print("GPUs found and memory growth enabled.")
    except RuntimeError as e:
        print(e)
else:
    print("No GPU found. Running on CPU.")

# ================================
# IMPORTS
# ================================
from tensorflow.keras import layers, Model
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
import time
import sklearn.metrics as skm
import seaborn as sns

# ================================
# DATA PREPARATION FOR KAGGLE
# ================================
data_dir = '/kaggle/input/breast-cancer-ultrasound-images/BUSI_aug'
img_size = 300  # Resize images to 300x300
batch_size = 8

train_ds = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    validation_split=0.3,
    subset='training',
    seed=42,
    image_size=(img_size, img_size),
    batch_size=batch_size,
    label_mode='categorical',
    interpolation='bicubic'
)
val_ds = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    validation_split=0.3,
    subset='validation',
    seed=42,
    image_size=(img_size, img_size),
    batch_size=batch_size,
    label_mode='categorical',
    interpolation='bicubic'
)

# Calculate class weights for imbalanced data.
class_names = train_ds.class_names
class_counts = np.bincount([label.numpy().argmax() for _, label in train_ds.unbatch()])
total_samples = class_counts.sum()
class_weights = {i: total_samples / (len(class_names) * class_counts[i]) for i in range(len(class_names))}

# ================================
# HVIT_AAF MODEL ARCHITECTURE (Data augmentation removed)
# ================================
def HVIT_AAF(input_shape=(300, 300, 3), num_classes=len(class_names)):
    base_model = tf.keras.applications.EfficientNetB0(
        include_top=False,
        weights='imagenet',
        input_shape=input_shape
    )
    base_model.trainable = True

    def mobilevit_block(x, num_blocks=4):
        for _ in range(num_blocks):
            residual = x
            x = layers.LayerNormalization()(x)
            x = layers.MultiHeadAttention(num_heads=2, key_dim=64)(x, x)
            x = layers.Add()([residual, x])
        return x

    inputs = layers.Input(shape=input_shape)
    x = inputs  # Data augmentation removed – input is directly passed in.
    x = base_model(x)
    cnn_features = layers.Conv2D(128, 3, padding='same')(x)
    # Reshape CNN features to feed into the mobilevit block.
    vit_features = layers.Reshape((-1, 128))(cnn_features)
    vit_features = mobilevit_block(vit_features)
    # Adjust the reshape target to match the actual spatial size (10x10 = 100)
    vit_features = layers.Reshape((10, 10, 128))(vit_features)
    attention = layers.Concatenate()([cnn_features, vit_features])
    attention = layers.Conv2D(1, 1, activation='sigmoid')(attention)
    fused = layers.Multiply()([cnn_features, attention])
    x = layers.GlobalAvgPool2D()(fused)
    x = layers.Dense(64, activation='swish')(x)
    x = layers.Dropout(0.5)(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)
    return Model(inputs, outputs)

# ================================
# CUSTOM LOSS FUNCTION
# ================================
class AdaptiveFocalLoss(tf.keras.losses.Loss):
    def __init__(self, alpha=0.25, gamma=2.0):
        super(AdaptiveFocalLoss, self).__init__()
        self.alpha = alpha
        self.gamma = gamma
        
    def call(self, y_true, y_pred):
        ce_loss = tf.keras.losses.categorical_crossentropy(y_true, y_pred)
        pt = tf.exp(-ce_loss)
        focal_loss = self.alpha * (1 - pt)**self.gamma * ce_loss
        return tf.reduce_mean(focal_loss)

# ================================
# GRAD-CAM FUNCTIONS
# ================================
def make_gradcam_heatmap(img_array, model, last_conv_layer_name, pred_index=None):
    grad_model = tf.keras.models.Model([model.inputs],
                                       [model.get_layer(last_conv_layer_name).output, model.output])
    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_array)
        if pred_index is None:
            pred_index = tf.argmax(predictions[0])
        loss = predictions[:, pred_index]
    grads = tape.gradient(loss, conv_outputs)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    conv_outputs = conv_outputs[0]
    heatmap = conv_outputs @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    heatmap = tf.maximum(heatmap, 0) / (tf.math.reduce_max(heatmap) + tf.keras.backend.epsilon())
    return heatmap.numpy()

def overlay_heatmap(img, heatmap, alpha=0.4, colormap=cv2.COLORMAP_JET):
    heatmap = np.uint8(255 * heatmap)
    heatmap = cv2.applyColorMap(heatmap, colormap)
    output = cv2.addWeighted(img, alpha, heatmap, 1 - alpha, 0)
    return output

# ================================
# MODEL COMPILATION
# ================================
def compile_model(model):
    model.compile(
        optimizer=tf.keras.optimizers.AdamW(learning_rate=1e-4),
        loss=AdaptiveFocalLoss(),
        metrics=[
            'accuracy',
            tf.keras.metrics.AUC(name='auc'),
            tf.keras.metrics.Precision(name='precision'),
            tf.keras.metrics.Recall(name='recall')
        ]
    )
    return model

# ================================
# TRAINING MULTIPLE RUNS WITH COMPUTATIONAL EFFICIENCY & COMBINED METRIC SELECTION
# ================================
num_runs = 3
epochs_per_run = 25
all_histories = []
best_combined_metric = -np.inf  # Combined metric to choose the best model
best_model = None

for run in range(num_runs):
    print(f"Training Run {run+1}/{num_runs}")
    start_time = time.time()
    model = HVIT_AAF()  # Using input shape 300x300
    model = compile_model(model)
    run_callbacks = [
        tf.keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True),
        tf.keras.callbacks.ReduceLROnPlateau(factor=0.5, patience=5)
    ]
    history = model.fit(
        train_ds,
        validation_data=val_ds,
        epochs=epochs_per_run,
        callbacks=run_callbacks,
        class_weight=class_weights
    )
    training_time = time.time() - start_time
    print(f"Training time for Run {run+1}: {training_time:.2f} seconds")
    all_histories.append(history)
    results = model.evaluate(val_ds, verbose=0)
    # results order: [loss, accuracy, auc, precision, recall]
    combined_metric = results[1] + results[2] + results[3] + results[4] - results[0]
    print(f"Run {run+1} metrics: Loss: {results[0]:.4f}, Accuracy: {results[1]:.4f}, AUC: {results[2]:.4f}, Precision: {results[3]:.4f}, Recall: {results[4]:.4f}, Combined Metric: {combined_metric:.4f}")
    if combined_metric > best_combined_metric:
        best_combined_metric = combined_metric
        best_model = model
        # Save the best model in HDF5 format, which is ideal for further conversion/integration.
        best_model.save('/kaggle/working/best_model.h5')



# ================================
# PLOTTING FUNCTIONS FOR TRAINING METRICS
# ================================
def plot_train_metric(metric, histories):
    plt.figure(figsize=(8, 6))
    for i, history in enumerate(histories):
        plt.plot(history.history[metric], label=f'Run {i+1}')
    plt.xlabel('Epoch')
    plt.ylabel(metric)
    plt.title(f'Training {metric} Across Runs')
    plt.legend()
    plt.show()

def plot_val_metric(metric, histories):
    plt.figure(figsize=(8, 6))
    for i, history in enumerate(histories):
        plt.plot(history.history[f'val_{metric}'], label=f'Run {i+1}')
    plt.xlabel('Epoch')
    plt.ylabel(metric)
    plt.title(f'Validation {metric} Across Runs')
    plt.legend()
    plt.show()

for met in ['accuracy', 'auc', 'precision', 'recall', 'loss']:
    plot_train_metric(met, all_histories)
    plot_val_metric(met, all_histories)

# ================================
# PREDICTION AND GRAD-CAM VISUALIZATION
# ================================
test_ds = val_ds.take(1)

for images, labels in test_ds:
    predictions = best_model.predict(images)
    for i in range(len(images)):
        pred_label = np.argmax(predictions[i])
        true_label = np.argmax(labels[i].numpy())
        # Retrieve the last Conv2D layer name for Grad-CAM.
        last_conv_layer_name = None
        for layer in best_model.layers[::-1]:
            if isinstance(layer, tf.keras.layers.Conv2D):
                last_conv_layer_name = layer.name
                break
        heatmap = make_gradcam_heatmap(np.expand_dims(images[i], axis=0),
                                       best_model, last_conv_layer_name)
        img_np = images[i].numpy().astype(np.uint8)
        heatmap_resized = cv2.resize(heatmap, (img_np.shape[1], img_np.shape[0]))
        superimposed_img = overlay_heatmap(img_np, heatmap_resized)
        plt.figure(figsize=(10, 4))
        plt.subplot(1, 2, 1)
        plt.imshow(img_np)
        plt.title(f"True: {class_names[true_label]}, Pred: {class_names[pred_label]}")
        plt.axis('off')
        plt.subplot(1, 2, 2)
        plt.imshow(cv2.cvtColor(superimposed_img, cv2.COLOR_BGR2RGB))
        plt.title("Grad-CAM\n(Prediction Influencing Area)")
        plt.axis('off')
        plt.show()

# ================================
# EVALUATION METRICS: CONFUSION MATRIX, ROC, AND PRECISION-RECALL CURVES
# ================================
# Collect predictions, true labels, and original images from the entire validation set.
y_true = []
y_pred = []
y_score = []
images_list = []
for images, labels in val_ds:
    preds = best_model.predict(images)
    y_score.extend(preds)
    y_true.extend(labels.numpy())
    y_pred.extend(np.argmax(preds, axis=1))
    for img in images:
        images_list.append(img.numpy())

y_true = np.array(y_true)  # one-hot encoded true labels
y_pred = np.array(y_pred)
y_score = np.array(y_score)

# --- 1. Confusion Matrix ---
# Convert one-hot true labels to integer labels.
cm = skm.confusion_matrix(np.argmax(y_true, axis=1), y_pred)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()

# --- 2. ROC Curve and AUC ---
n_classes = len(class_names)
plt.figure(figsize=(8, 6))
for i in range(n_classes):
    fpr, tpr, _ = skm.roc_curve(y_true[:, i], y_score[:, i])
    roc_auc = skm.auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'{class_names[i]} (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Each Class')
plt.legend()
plt.show()

# --- 3. Precision-Recall Curve ---
plt.figure(figsize=(8, 6))
for i in range(n_classes):
    precision, recall, _ = skm.precision_recall_curve(y_true[:, i], y_score[:, i])
    plt.plot(recall, precision, label=class_names[i])
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve for Each Class')
plt.legend()
plt.show()

# ================================
# ERROR ANALYSIS: Misclassified Samples
# ================================
misclassified_indices = np.where(np.argmax(y_true, axis=1) != y_pred)[0]
num_to_plot = min(5, len(misclassified_indices))
plt.figure(figsize=(15, 3))
for i, idx in enumerate(misclassified_indices[:num_to_plot]):
    plt.subplot(1, num_to_plot, i+1)
    plt.imshow(images_list[idx].astype('uint8'))
    true_label = class_names[np.argmax(y_true[idx])]
    pred_label = class_names[y_pred[idx]]
    plt.title(f"True: {true_label}\nPred: {pred_label}")
    plt.axis('off')
plt.suptitle("Misclassified Samples")
plt.show()

# ================================
# COMPUTATIONAL EFFICIENCY: Inference Timing
# ================================
# Measure inference time per image on one batch.
for images, labels in val_ds.take(1):
    start = time.time()
    _ = best_model.predict(images)
    end = time.time()
    inference_time = (end - start) / images.shape[0]
    print(f"Inference time per image: {inference_time:.4f} seconds")

# ================================
# FINAL MODEL PERFORMANCE SUMMARY
# ================================
print("Best Model Evaluation Metrics on Validation Set:")
results = best_model.evaluate(val_ds, verbose=0)
print(f"Loss: {results[0]:.4f}")
print(f"Accuracy: {results[1]:.4f}")
print(f"AUC: {results[2]:.4f}")
print(f"Precision: {results[3]:.4f}")
print(f"Recall: {results[4]:.4f}")

# Note:
# For integrating the saved model with a frontend (for example, in a web application),
# you can convert the HDF5 model to the TensorFlow.js format using the command:
#
#   tensorflowjs_converter --input_format=keras /kaggle/working/best_model.h5 /path/to/web_model
#
# This conversion produces a set of JSON and binary weight files that can be loaded directly
# in the browser using TensorFlow.js for fast image classification.
```
